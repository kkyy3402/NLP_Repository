### concat과 merge는 다르다.

```python
df2 = pd.read_csv("korean-idol2.csv")
```

### merge 는 unique 키를 기준으로 합쳐주는 기능을 한다.

```python
df2
```

```python
pd.merge(df,df2,on="이름", how="right") # right 기준으로 있는 값들만 남기고 다 drop이 된다.
```

### how 옵션 : left, right, inner(교집합), outer (합집합)

```python
pd.merge(df,df2,on="이름",how="inner")
```

pd.merge(df,df2, left_on="이름", right_on="성함", how="outer")  
이런식으로 쓰면, 데이터 프레임간의 컬럼 명이 달라도 커버가 가능하다.
