### Apply

```python
import pandas as pd
```

```python
df = pd.read_csv("korean-idol.csv")
```

```python
df
```

```python
df.loc[df['성별'] == "남자"]
```

### apply 함수 ㄱ ㄱ 

```python
def male_or_female(x):
    if x == "남자":
        return 0
    else :
        return 1
    
df["성별"].apply(male_or_female)
```

### apply를 한번 더 써보자

```python
def cm_to_brand(df):
    value = df["브랜드평판지수"] / df["키"]
    return value

df.apply(cm_to_brand, axis=1)
```

### lambda로 apply 해보자

```python
df["성별"].apply(lambda x: 1 if x =="남자" else 0)
```

### map을 통한 데이터 조작

```python
my_map = {
    "남자" : 1, 
    "여자" : 0
}

df["성별"].map(my_map)
```
