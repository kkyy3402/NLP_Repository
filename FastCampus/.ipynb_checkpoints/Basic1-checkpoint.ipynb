## 범위 선택

### 단순 index에 대한 범위 선택 

```python
df[:3]
```

```python
df.head(3)
```

### loc를 통한 범위 조회

```python
df.loc[:,"Stage Name"]
```

### 이런식으로 여러 컬럼의 내용을 가져올 수 있다. 

```python
df.loc[:,["Stage Name","Full Name"]]
```

```python
df.head()
```

### 2번째부터 5번쨰까지 (기존 파이썬이랑 인덱스 기준이 다르다)

```python
df.loc[2:5,["Full Name","Group"]]
```

### iloc : 컬럼 인덱스를 통한 조회

```python
df.iloc[2:5,[1,4]]
```

### Boolean Indexing - 조건을 활용한 색인  
예를 들어 키가 몇 이상인 사람만 조회하고 싶을 때 사용 

### 예를 들어 Gender가 M인 사람만 가져오고 싶을 때..

```python
df[df["Gender"] == "M"]
```

### 조건 + 특정 컬럼

```python
df.loc[df["Gender"] == "M", ["Group","Gender"]]
```

### isin ( ...가 ...에 있어 )
예를 들어 BAP라는 그룹과 SF9라는 그룹에 있는 사람만 뽑아오고 싶을땐.. 

```python
my_condition = ["B.A.P","SF9"]
df[df["Group"].isin(my_condition)]
```

### Nan 값만 조회하기
